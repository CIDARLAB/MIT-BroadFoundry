```python
blasts = open("../../../int_xis/mining/int_xis_nuc_bacterial_genome_BLAST.bl")
```

```python
int_xis = open("../../../int_xis/mining/int_xis_protein.fasta")
```

```python
seqs = dict()
line1 = int_xis.readline()
line2 = int_xis.readline()
while not line1 == "":
    if not "xis" in line1.lower().split(" ")[1]:
        seqs[line1[1:-1]] = line2.strip()
    line1 = int_xis.readline()
    line2 = int_xis.readline()
```

```python
seqs
```

```python
fasta = []
query = 0
for line in blasts:
    line = line.strip()
    if line.startswith("# Query: "):
        query = 1
        if len(fasta) > 1:
            f = open("../../../int_xis/mining/alignments/" + intxis[:50].replace("/", "-") + ".fasta", "w")
            f.write("".join(fasta))
            f.close()
        intxis = line.replace("# Query: ", "")
            
        fasta = []
        
        if intxis not in seqs:
            query = 0
            continue
        fasta.append(">{}\n{}\n".format(intxis, seqs[intxis]))
    elif query == 0:
        continue
    elif line.startswith("#"):
        continue
    else:
        acc = line.split("\t")[1].split("|")[1]
        try:
            phage = open("../../../int_xis/{}_detail.fas".format(acc))
        except:
            continue
        prev = 0
        for line in phage:
            if "integrase" in line:
                fasta.append(line)
                prev = 1
            elif prev == 1:
                fasta.append(line)
                prev = 0
        
```
