```python
import subprocess
```

```python
detail_files = [f for f in subprocess.check_output(["ls", "../../../int_xis/"]).split("\n") if "_detail" in f]
```

```python
detail_files
```

```python
for f in detail_files:
    g = []
    fi = open("../../../int_xis/" + f)
    for line in fi:
        line = line.replace("  ", "\t")
        line = line.replace("\t ", "\t")
        while line.find("\t\t") > 0:
            line = line.replace("\t\t", "\t")
        if ("integrase" in line or "att" in line) and len(line.strip().split("\t")) == 4:
            rge, hit, e, seq = line.strip().split("\t")
            g.append(">{}|{}|{}\n{}\n".format(hit, e, rge, seq))
    fi.close()
    if len(g) < 2:
        continue
    gf = open("../../../int_xis/" + f.split(".")[0] + ".fas", "w")
    gf.write("".join(g))
    gf.close()
```
