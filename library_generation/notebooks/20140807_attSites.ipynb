```python
import subprocess
```

```python
detail_files = [f for f in subprocess.check_output(["ls", "../../../int_xis/"]).split("\n") if "_detail.fas" in f]
```

```python
detail_files
```

```python
site_dict = dict()
for f in detail_files:
    f = open("../../../int_xis/" + f)
    attl = []
    attr = []
    integrases = dict()
    for seq in f.read().strip(">").split("\n>"):
        if 'attL' in seq:    
            attl.append(seq.split("\n")[1])
        if 'attR' in seq:
            attr.append(seq.split("\n")[1])
        if 'integrase' in seq:
            integrases[seq.split("\n")[1]] = attl
        if len(attr) == len(attl) and not len(attr) == 0:
            for integrase in integrases:
                site_dict[integrase] = integrases[integrase]
            integrases = dict()
```
