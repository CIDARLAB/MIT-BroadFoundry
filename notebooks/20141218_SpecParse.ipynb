```python
import itertools
```

```python
import re
```

```python
def parens(spec):
    if len([c for c in spec if c in "()+*_~"]) == 0:
        return [spec]
    new = []
    prev_counter = 0
    counter = 0
    save = True
    for char in re.split("([()+*_~])", spec):
        if not char:
            continue
        if char == "(":
            counter += 1
            if prev_counter == 0:
                save = False
                new.append([])
        if char == ")":
            counter -= 1
            if prev_counter == 1:
                save = False
                
        if counter == 0 and prev_counter == 0:
            new.append(char)
        else:
            if save:
                new[-1].append(char)
            
        if not save:
            save = True
        prev_counter = counter
    return ["".join(n) for n in new]
```

```python
def get(thing):
    stuff = {'A':[['X'], ['Y']], 'B': [['Q']], 'C': [['J'], ['K']], 'D': [['p'], ['u']], 'E': [['g'], ['h']], 'F': [['Z'], ['V']]}
    return stuff[thing]
```

```python
def combine(group):
    combs = [list(itertools.chain(*c)) for c in itertools.product(*group)]
    return combs
```

```python
def star(group):
    perms = [sub for group in list(itertools.permutations([spec_parse(list(g)[0]) for k,g in itertools.groupby(group,
                                                                lambda x: x == '*') if not k])) for sub in combine(group)]
    return perms
```

```python
def plus(group):
    combs = combine([spec_parse(list(g)[0]) for k,g in itertools.groupby(group, lambda x: x == '+') if not k])
    return combs
```

```python
def spec_parse(spec):
    separated = parens(spec)
    if '*' in separated:
        return star(separated)
    elif '+' in separated:
        return plus(separated)
    elif "(" in separated[0] or '*' in separated[0] or '+' in separated[0]:
        return spec_parse(separated[0])
    else:
        return get(separated[0])
```
